# Testing hypotheses

```python
# Load the data from the CSV file
df_rides = pd.read_csv("/datasets/project_sql_result_07.csv")

# Display the first few rows of the DataFrame
print(df_rides.head())

# Display DataFrame information
print(df_rides.info())

```

```python
# Group the data by weather_conditions and calculate the average duration_seconds for each group
average_duration_by_weather = df_rides.groupby('weather_conditions')['duration_seconds'].mean()

# Display the average duration for each group
print("Average ride duration by weather conditions:")
print(average_duration_by_weather)
```

```python
import numpy as np
import scipy.stats as stats


# Convert 'start_ts' column to datetime type
df_rides['start_ts'] = pd.to_datetime(df_rides['start_ts'])

# Separate the data into two groups: rides on rainy Saturdays and rides on non-rainy Saturdays
rainy_saturdays = df_rides[(df_rides['weather_conditions'] == 'Bad') & (df_rides['start_ts'].dt.dayofweek == 5)]
non_rainy_saturdays = df_rides[(df_rides['weather_conditions'] == 'Good') & (df_rides['start_ts'].dt.dayofweek == 5)]

# Perform the hypothesis test
t_stat, p_value = stats.ttest_ind(rainy_saturdays['duration_seconds'], non_rainy_saturdays['duration_seconds'])

# Set the significance level
alpha = 0.05

# Print the results
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant difference in the average duration of rides between rainy and non-rainy Saturdays.")
else:
    print("Fail to reject the null hypothesis. There is no statistically significant difference in the average duration of rides between rainy and non-rainy Saturdays.")
```

Null Hypothesis:
The average duration of rides from the Loop to O'Hare International Airport is the same on rainy Saturdays as it is on non-rainy Saturdays.
Alternative Hypothesis:
The average duration of rides from the Loop to O'Hare International Airport is different on rainy Saturdays compared to non-rainy Saturdays.
We compared the observed data (in this case, the duration of rides on rainy and non-rainy Saturdays) and calculated the probability of observing this data under the assumption that the null hypothesis is true. If this probability (p-value) is below a pre-defined significance level, typically set at 0.05, we reject the null hypothesis and conclude that there is a statistically significant difference between the groups.

# Conclusion

To test these hypotheses, we need a statistical method that can compare the means of two independent groups. The data consists of continuous numerical values representing the duration of rides, which is suitable for comparing means. The statistical test we choose should meet certain assumptions about the data. We used a two-sample t-test. We set a significance level to determine the threshold for statistical significance. Commonly is set to 0.05, indicating a 5% chance of making a Type I error (rejecting the null hypothesis when it is true). In summary, we chose a statistical criterion based on comparing means because it aligns with the nature of the hypotheses, the type of data, and the assumptions of the statistical tests available. 

In this analysis, we aimed to test two hypotheses related to taxi rides in Chicago. The first hypothesis focused on examining whether the average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays. The second hypothesis aimed to understand the impact of weather conditions on ride frequency for different taxi companies.

To test these hypotheses, we employed statistical methods that could compare means of two independent groups. The data provided consisted of continuous numerical values representing the duration of rides, making it suitable for comparing means. We selected the two-sample t-test as our statistical test, as it aligns with the nature of the hypotheses, the type of data available, and the assumptions of the statistical tests.

For the first hypothesis, we categorized the data into two groups: rides on rainy Saturdays and rides on non-rainy Saturdays. We then compared the average duration of rides between these two groups using the two-sample t-test.

For the second hypothesis, we analyzed the relationship between weather conditions and ride frequency for different taxi companies. By examining the number of rides for each company under different weather conditions, we gained insights into passenger preferences and the impact of external factors on ride demand.
